```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../partition_data.csv")
df = df.rename(columns={'partitionKey': 'Partition Key', 'count': 'Count', 'experiment': 'Experiment', 'size': 'Data Size'})

df_small = df[df['Data Size'] == '15 min']
df_large = df[df['Data Size'] == '6 hrs']
df.head()
```

```python
plt.figure(figsize=(15, 5))

def plot_experiment(df, category, size_small='15 min', size_large='6 hrs'):
    # Filter data
    reg = df[df['Experiment'] == category]
    reg['Mean Value'] = reg['Data Size'].map(reg.groupby('Data Size')['Count'].mean())
    reg['Normalized Count'] = reg.groupby('Data Size')['Count'].transform(lambda x: (x - x.min()) / (x.max() - x.min()))
    reg_small = reg[reg['Data Size'] == size_small]
    reg_large = reg[reg['Data Size'] == size_large]
    # Plot and save the small size data
    plt.figure(figsize=(15, 5))
    sns.barplot(data=reg, x="Partition Key", y='Count', hue='Data Size')
    sns.lineplot(data=reg, x="Partition Key", y='Mean Value', hue='Data Size')
    plt.title(f'{category} - Distribution Bar Plot')
    plt.savefig(f'{category}_barplot.png')
    plt.show()
    plt.close()

    # Plot and save the large size data
    plt.figure(figsize=(15, 5))
    sns.displot(reg, x='Partition Key', y='Count' , hue='Data Size')
    plt.title(f'{category} - Normalized Counts per Tile Histogram')
    plt.savefig(f'{category}_distplot.png')
    plt.show()
    plt.close()
```

```python
for exp in df['Experiment'].unique():
    plot_experiment(df, exp)
```

```python
from math import log, log10

exp = pd.concat([exp1, exp2, exp3, exp4])
exp['log01'] = exp['0.1'].astype(float).apply(lambda x: log10(x))
exp['log02'] = exp['0.2'].astype(float).apply(lambda x: log10(x))
exp['log05'] = exp['0.5'].astype(float).apply(lambda x: log10(x))

def normalize(x):
    return (x - x.min()) / (x.max() - x.min())

normalized_df = exp.copy()
for col in ['log01', 'log02', 'log05']:
    normalized_df[col] = normalized_df.groupby('Query Number')[col].transform(normalize)

normalized_df.head()
```

```python
def plot_log_scale(df, log_scale, scale_factor, query_number):
    df_long = pd.melt(df, id_vars=['Query Number', 'Exp'], value_vars=[log_scale], 
                      var_name='Log Scale', value_name='Performance')
    
    plt.figure(figsize=(15, 5))
    sns.barplot(data=df_long, x='Query Number', y='Performance', hue='Exp')
    plt.title(f'Query Log Time for Scale Factor {scale_factor}')
    plt.savefig(f'{log_scale}_barplot.png')
    plt.show()

# Plot for log01
plot_log_scale(exp, 'log02', 0.2, 3)
```
